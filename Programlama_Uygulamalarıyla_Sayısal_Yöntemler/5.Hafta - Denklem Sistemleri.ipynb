## Gauss Jordan Yöntemi 

Burada amaç alt/üst üçgensel bir matris oluşturmak yerine, elementer işlemler sonucunda köşegen veya birim matris oluşturulması amaçlanmaktadır.



Burada 3 bilinmeyenli 3 denklem yer almaktadır ve bu yöntem sayesinde bilinmeyenler rahatlıkla bulunabilmektedir.

#### Örn : 

x+ 2y +z = 8

2x -y + z = 3 

3x + 3y -2z = 3 denklem sistemindeki bilinmeyenleri program yazarak bulunuz. 


```python
import numpy as np 
```

```python
A = np.matrix('1.0 2.0 1.0 ; 2.0 -1.0 1.0 ; 3.0 3.0 -2.0')
b = np.matrix('8.0;3.0;3.0')
n = len(b)
```

```python
for colm in range(n):
    for row in range(n):
        if colm != row:
            c = -(A[row,colm] / A[colm, colm])
            A[row] = np.multiply(c,A[colm]) + A[row]
            b[row] = c*b[colm] + b[row]
            
print(A)
print(b)

for i in range(n):
    x[i] = (b[i]/ A[i,i])
print(x)
             
```
